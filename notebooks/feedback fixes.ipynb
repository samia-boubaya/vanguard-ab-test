### STEP 02: Merge & Save in CSV files

```python
import pandas as pd

# Save each part as CSV
final_web_data_pt_1.to_csv("../data/processed/final_web_data_pt_1.csv",index=False)
final_web_data_pt_2.to_csv("../data/processed/final_web_data_pt_2.csv",index=False)

# Merge the two parts (same columns)
final_web_data = pd.concat([final_web_data_pt_1, final_web_data_pt_2],ignore_index=True)

# Save merged CSV
final_web_data.to_csv("../data/processed/final_web_data_merged.csv",index=False)

# Display the merged DataFrame to ensure it loaded correctly
display(final_web_data.head())
```
