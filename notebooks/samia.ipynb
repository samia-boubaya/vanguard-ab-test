### Merging dfs

```python
import pandas as pd

# Load the combined dataset
df = pd.read_csv('../data/df_final_web_data.csv')

# Convert date_time to datetime type
df['date_time'] = pd.to_datetime(df['date_time'])

# Split into two columns
df['date'] = df['date_time'].dt.date
df['time'] = df['date_time'].dt.time

# drop the original date_time column
df = df.drop(columns=['date_time'])

# Save the updated DataFrame
df.to_csv('../data/df_final_web_data.csv', index=False)
print("Date and time columns added and saved in 'df_final_web_data.csv'")
```
