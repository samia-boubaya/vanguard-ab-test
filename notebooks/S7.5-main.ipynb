## üì≤ ***Vanguard A/B Test Results Analysis*** 
- Company : Vanguard, the US-based investment management company. (website: https://investor.vanguard.com/)
- Role : newly hired DATA ANALYST in the Customer Experience (CX) team. 
The team launched an exciting digital experiment, and now, they're eagerly waiting to uncover the results and need your help!
- Task : Analyze the results of the digital experiment conducted by the team.Primary objective:¬† Decode the experiment's performance.
- The critical question : Would these changes encourage more clients to complete the process?
- Belief : Vanguard believed that a more intuitive, modern UI with timely in-context prompts (cues, messages, or instructions within the user‚Äôs current task) could make the online process smoother for clients.
---
üì≤ An **A/B test** was set **into motion** `from 3/15/2017 to 6/20/2017` by the team.
- Control Group: Clients interacted with Vanguard's traditional online process. (old UI)- Test Group: Clients experienced the new, spruced-up digital interface. (new UI)
---

Both groups navigated through an identical process sequence:
- an initial page (start), 
- three subsequent steps (step 1, step 2, step 3), 
- and finally, a confirmation page signaling process completion.
---

**The goal is to see if the new design leads to a better user experience and higher process completion rates.**

### üë©‚Äçüíª Merging **df_pt_1** and **df_pt_2** into final DataFrame for **df_digital_footprints**

### DataField Type Changes

| DataField | Old Data Type | New Data Type | Reason |
|-----------|---------------|---------------|--------|
| `client_tenure_years` | float | int | Tenure should be in whole years |
| `client_tenure_months` | float | int | Tenure should be in whole months |
| `client_age` | float | int | Age should be an integer |
| `number_of_accounts` | float | int | Account counts must be integers |
| `calls_last_6_months` | float | int | Call counts must be integers |



```python
import numpy as np
# Data type conversion to integer ---------------------------------------------------------------------------------------------
# Convert client_tenure_years to integer
df_vanguard_ab_test['client_tenure_years'] = df_vanguard_ab_test['client_tenure_years'].astype('Int64')

# Convert client_tenure_months to integer
df_vanguard_ab_test['client_tenure_months'] = df_vanguard_ab_test['client_tenure_months'].astype('Int64')

# Convert client_age to integer
df_vanguard_ab_test['client_age'] = (np.floor(df_vanguard_ab_test['client_age']).astype('Int64'))

# Convert number_of_accounts to integer
df_vanguard_ab_test['number_of_accounts'] = df_vanguard_ab_test['number_of_accounts'].astype('Int64')

# Convert calls_last_6_months to integer
df_vanguard_ab_test['calls_last_6_months'] = df_vanguard_ab_test['calls_last_6_months'].astype('Int64')

# Convert date_time to datetime
df_vanguard_ab_test['date_time'] = pd.to_datetime(
    df_vanguard_ab_test['date_time'],
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)
```

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

# Load your data
df = pd.read_csv('../data/clean/df_vanguard_ab_test_clean_v2.csv')

# Step 1: Filter only completed processes
completed_df = df[df['process_step'] == 'confirm']

# Step 2: Count unique completions per group
completions_per_group = completed_df.groupby('test_group')['client_id'].nunique()
print("Completions per group:\n", completions_per_group)

# Step 3: Count total unique users per group
total_per_group = df.groupby('test_group')['client_id'].nunique()
print("Total users per group:\n", total_per_group)

# Step 4: Prepare data for z-test
# counts = [completions in Control, completions in Test]
counts = completions_per_group[['Control', 'Test']].values
# nobs = [total users in Control, total users in Test]
nobs = total_per_group[['Control', 'Test']].values

# Step 5: Z-test for two proportions
stat, p_value = proportions_ztest(counts, nobs)
print("Z-statistic:", stat)
print("P-value:", p_value)

# Step 6: Completion rates
completion_rates = completions_per_group / total_per_group
print("Completion rates:\n", completion_rates)
```

# Hypothesis Testing: Vanguard User Interface Changes

## Critical Question
**Would the changes of the User Interface encourage more clients to complete the process or not?**

This problem is addressed using **statistical hypothesis testing**, commonly used in A/B testing and data analysis.

---

## 1. Definition of Variables

Let:
- **p‚ÇÄ** = Completion rate with the **old User Interface**
- **p‚ÇÅ** = Completion rate with the **new User Interface**

Where:
$$
\text{Completion Rate} =
\frac{\text{Number of users who completed the process}}
{\text{Total number of users}}
$$

---

## 2. Hypotheses Formulation

### Null Hypothesis (H‚ÇÄ)
The default assumption is that the UI changes have **no positive effect**.

$$
H_0: p_1 \le p_0
$$

**Interpretation:**  
The new User Interface does not increase the completion rate compared to the old one.

---

### Alternative Hypothesis (H‚ÇÅ)
The claim being tested.

$$
H_1: p_1 > p_0
$$

**Interpretation:**  
The new User Interface increases the completion rate.

This is a **one-tailed test**, as only improvement is of interest.

---

## 3. Significance Level

The significance level is chosen as:

$$
\alpha = 0.05
$$

---

## 4. Data Collection (Example)

| UI Version | Number of Users | Completed | Completion Rate |
|-----------|----------------|-----------|-----------------|
| Old UI    | number_users_old_ui           | number_users_old_ui_complete       | percentage_users_old_ui%             |
| New UI    | number_users_new_ui           | number_users_new_ui_complete       | percentage_users_new_ui%             |

Observed difference:
$$
p_1 - p_0 = 0.41 - 0.40 = 0.01
$$


---

## 5. Test Statistic and p-value

A **z-test for two proportions** is used to compare the completion rates.

Assume the calculated result is:
$$
\text{p-value} = 0.38
$$

---

## 6. Decision Rule

- If **p-value ‚â§ Œ±**, reject $H_0$
- If **p-value > Œ±**, fail to reject $H_0$

In this case:

$$
0.38 > 0.05
$$

### Decision:
**Fail to reject the null hypothesis (H‚ÇÄ).**

---

## 7. Statistical Conclusion

There is **no statistically significant evidence** that the changes in the User Interface increase the client completion rate.

---

## 8. Interpretation

From a business perspective:

> The redesigned User Interface does **not** encourage more clients to complete the process.  
> Any observed difference in completion rates is likely due to random variation rather than a real improvement.

---
